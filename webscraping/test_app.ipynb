```python
def move_xls_to_folder(destiny_folder, file_path):

    downloads_path = "C:/Users/javie/Downloads"

    for file in os.listdir(downloads_path):
        if file == "Tabla.zip":
            file_path = os.path.join(downloads_path, file) 
            with zipfile.ZipFile(file_path, 'r') as zip_ref:
                zip_ref.extractall(destiny_folder)
            
            if os.path.exists(file_path):
                os.remove(file_path)
                while os.path.exists(file_path):
                    time.sleep(1)     
            else:
                print("not exists") 
            break

    with open(file_path, 'r') as file:
        html_content = file.read()

    html_buffer = StringIO(html_content)

    df = pd.read_html(html_buffer)

    return df[0]
```

```python
general_path = path + "/data/general_info"
file_path = 'data/general_info/Contenido-tabla.xls'
df_general_info = move_xls_to_folder(general_path, file_path)
```
